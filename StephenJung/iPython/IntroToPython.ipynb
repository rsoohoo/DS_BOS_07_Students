```python
class CreditCard(object):
    def __init__(self, owner_name, exp_month, exp_year,
                 credit_limit, interest_rate, bal=0):
        """
        Initialize the credit care with the following values:
        owner_name,
        exp_month,
        exp_year,
        credit_limit,
        interest_rate,
        bal
        """
        self.owner_name=owner_name
        self.exp_month=exp_month
        self.exp_year=exp_year
        self.credit_limit=credit_limit
        self.interest_rate=interest_rate
        self.bal=bal
        
    def make_purchase(self, purchase_amount):
        """
        Update the balance by adding the purchase_amount
        """
        self.remaining_bal = self.credit_limit-self.balance
        if self.remaining_bal < purchase_amount:
            print ("Overdraw warning: p_amount {}, \
                   remaining_bal {}", purchase_amount, 
                   self.remaining_bal)
        else:
            self.bal+=purchase_amount
            print self.bal
        return
    
    def  make_payment(self, payment_amount):
        """
        Reduce the balance by the payment_amount
        """
        self.remaining_balance = self.credit_limit-self.balance
        if payment_amount > self.remaining_balance:
            print ("Warning: balance {}, payment_amount {}",
                  self.remaining_balance, payment_amount)
        else:
            self.bal-=payment_amount
        return
    
```
