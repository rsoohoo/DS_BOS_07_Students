#### Create a function that takes in an author and title string, and calls the correct things to generate a list of dicts containing its history


```python
from datetime import datetime as dt
def get_history(author, title, current_date):
    #Change strings to interact with the API
    author = author.encode('utf-8')
    title = title.encode('utf-8')
    if ' ' in author:
        author=author.replace(" ","%20")
    if ' ' in title:
        title=title.replace(" ","%20")
    if '&' in author:
        author=author.replace("&","%26")
    if '&' in title:
        title=title.replace("&","%26")
    if '(' in author:
        author=author.replace("(","%28")
    if '(' in title:
        title=title.replace("(","%28")
    if ')' in author:
        author=author.replace(")","%29")
    if ')' in title:
        title=title.replace(")","%29")
    if "'" in author:
        author=author.replace("'","%27")
    if "'" in title:
        title=title.replace("'","%27")
    request_string = 'http://api.nytimes.com/svc/books/v2/lists/best-sellers/history.json?author='+author+'&title='+title+'&api-key='+books_key
    response = urllib2.urlopen(request_string)
    data = json.load(response)
    results = data['results']
    history_on_list = {}
    current_date = dt.strptime(current_date, "%Y-%m-%d")
    if len(results) > 0:
        for historic_record in results[0]['ranks_history']:
            # Find the date this history record references, to make sure it doesn't reference something before our scaled back day
            this_date = dt.strptime(historic_record['published_date'], "%Y-%m-%d")
            if (historic_record['display_name'] == 'Combined Print & E-Book Fiction' and (current_date - this_date).days > 0):
                history_on_list[(current_date - this_date).days/7] = historic_record['rank']
    return history_on_list
print get_history('Kristin Hannah', 'THE NIGHTINGALE', current_published)
```

## Function for building desired data frame given results

```python
def make_df(results):
    data_for_df=[]
    current_date = results[0]['published_date']
    for book in results:
        history = get_history(book['book_details'][0]['author'], book['book_details'][0]['title'], current_date)
        current_dict = {'isbn': book['book_details'][0]['primary_isbn13'], 
                        'title': book['book_details'][0]['title'],
                        'author': book['book_details'][0]['author'], 
                        'publisher': book['book_details'][0]['publisher'],
                        'weeks': book['weeks_on_list'], 
                        'rank': book['rank'], 
                        'bulk': book['dagger'],
                        'date': book['published_date']}
        for i in range(1,11):
            if i in history:
                current_dict[i] = history[i]
            else:
                current_dict[i] = 0
        data_for_df.append(current_dict)
    return data_for_df
    
#For purposes of testing, extract one 3 weeks before the most recently available, so to test going forward
request_string = 'http://api.nytimes.com/svc/books/v2/lists/2015-12-20/combined-print-and-e-book-fiction.json?api-key='+books_key
response = urllib2.urlopen(request_string)
data = json.load(response)
results = data['results']
data_to_use = make_df(results)
print data_to_use
```
