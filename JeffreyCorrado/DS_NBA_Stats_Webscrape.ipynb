```python
from bs4 import BeautifulSoup
import urllib
from urllib.request import urlopen
page = urllib.request.urlopen('http://espn.go.com/nba/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2015/qualified/false/count/1').read()
scoring = BeautifulSoup(page)

import csv
import requests 


player_scoring_url = []
for i in range(1, 492, 40):
    url= "http://espn.go.com/nba/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2015/qualified/false/count/{}".format(i)
    r = requests.get(url)
    scoring = BeautifulSoup(r.content)
    player_scoring_url.append(scoring)
    
print(type(player_scoring_url))

print(len(player_scoring_url))


```

```python
scoring.prettify()[0:1000]
```

```python
scoring_div = scoring.find(id="my-players-table" )
print(scoring_div.prettify()[0:4000])
```

```python
scoring_table = scoring.find('table')
print(scoring_table)
```

```python
table_header = scoring_table.find(attrs={"class":'colhead'})
print(table_header.prettify())
```

```python
header_columns = table_header.findAll('td')
print(header_columns)
```

```python
cols = []
for header_column in header_columns:
    val = header_column.string
    if val != 'RK':
        cols.append(val)
    if val == 'PLAYER':
        cols.append('POSITION')
print(cols)
```

```python
scoring_rows = scoring_table.findAll('tr')
print(scoring_rows)
```

```python
players_scoring_dicts = []
players_scoring_array = []
for row in scoring_rows:
    if row.attrs['class'][0]=='colhead':
        continue
    player_scoring_stats = []
    scoring_row_cols = row.find_all('td')
    scoring_col_vals = []
    player_scoring_col = scoring_row_cols[1]
    player_name = player_scoring_col.find('a').string
    player_position = player_scoring_col.contents[1]
    player_position = player_position.split(' ')[1]
    player_scoring_stats.append(player_name)
    player_scoring_stats.append(player_position)
    for i in range(2, len(scoring_row_cols)):
        stat = scoring_row_cols[i].string
        player_scoring_stats.append(stat)
    players_scoring_array.append(player_scoring_stats)
    player_scoring_stats = dict(zip(cols, player_scoring_stats))
    players_scoring_dicts.append(player_scoring_stats)
print(players_scoring_dicts[0:4000])  
```
